# Script para gerar confrontos e calendário de forma aleatória

```python
import random

def gerar_calendario(times, horarios):
    # Lista de times e configuração básica
    lista_times = list(times.keys())
    n_times = len(lista_times)
    n_rodadas = n_times - 1
    jogos_por_rodada = n_times // 2

    # Shuffle para evitar padrão repetitivo
    random.shuffle(lista_times)

    # Geração de confrontos (Round-robin scheduling)
    confrontos = []
    for _ in range(n_rodadas):
        rodada_atual = []
        for i in range(jogos_por_rodada):
            time1 = lista_times[i]
            time2 = lista_times[-(i + 1)]
            rodada_atual.append((time1, time2))
        confrontos.append(rodada_atual)
        # Rotação dos times para a próxima rodada
        lista_times = [lista_times[0]] + lista_times[-1:] + lista_times[1:-1]

    # Separar confrontos em turnos
    primeiro_turno = []
    segundo_turno = []
    for rodada in confrontos:
        rodada_primeiro_turno = []
        rodada_segundo_turno = []
        for time1, time2 in rodada:
            # Primeiro turno: time1 em casa
            rodada_primeiro_turno.append({
                "mandante": time1,
                "visitante": time2,
                "estadio": times[time1]["estadio"]
            })
            # Segundo turno: time2 em casa
            rodada_segundo_turno.append({
                "mandante": time2,
                "visitante": time1,
                "estadio": times[time2]["estadio"]
            })
        primeiro_turno.append(rodada_primeiro_turno)
        segundo_turno.append(rodada_segundo_turno)

    # Alocar horários para os jogos
    horarios_ciclo = horarios * (len(primeiro_turno[0]) // len(horarios) + 1)
    for rodada in primeiro_turno + segundo_turno:
        random.shuffle(horarios_ciclo)  # Aleatorizar os horários
        for i, jogo in enumerate(rodada):
            jogo["horario"] = horarios_ciclo[i]


    return {"primeiro_turno": primeiro_turno, "segundo_turno": segundo_turno}

# Geração do calendário
calendario = gerar_calendario(times, horarios)

# Exibindo o calendário
for turno, rodadas in calendario.items():

    print(f"\n{turno.capitalize()}:")
    for i, rodada in enumerate(rodadas, 1):
        print(f"Rodada {i}:")
        for jogo in rodada:
            print(f"{jogo['mandante']} x {jogo['visitante']} - Estádio: {jogo['estadio']} - Horário: {jogo['horario']}")
```

# Script final utilizando simulated_annealing e uma função de custos para decidir os confrontos.

### A função de custo avalia a qualidade de um calendário, penalizando deslocamentos entre cidades (baseado nas distâncias normalizadas) e desequilíbrios entre os níveis dos times (quando o mandante é mais fraco que o visitante). Na seleção de uma solução vizinha, o algoritmo modifica ligeiramente o calendário atual e compara os custos. Se o custo da vizinha for menor, a solução é sempre aceita; caso contrário, pode ser aceita probabilisticamente, permitindo explorar soluções subótimas para escapar de mínimos locais. Isso busca encontrar um calendário mais eficiente ao longo do processo.

```python
import random
import math

# função de custo
def calcular_custo(calendario, times, distancias):
    custo = 0

    for _, rodadas in calendario.items():
      for rodada in rodadas:
        for jogo in rodada:
            mandante = jogo["mandante"]
            visitante = jogo["visitante"]
            cidade_mandante = times[mandante]["cidade"]
            cidade_visitante = times[visitante]["cidade"]

            # evitar problema dos times serem da mesma cidade
            if cidade_mandante != cidade_visitante:
                custo += distancias[cidade_mandante][cidade_visitante] / 100.0  # normalizar distâncias

            # confrontos desbalanceados (nível do mandante vs visitante)
            nvl_mandante = times[mandante]["nvl"]
            nvl_visitante = times[visitante]["nvl"]
            if nvl_mandante < nvl_visitante:  # quando o mandante é mais fraco
                custo += 1.0  # penalizando

    return custo

# recozimento simulado (algoritmo de otimização baseado no processo de choque térmico em metais)
def simulated_annealing(times, horarios, distancias, max_iter=1000, temp=1000, cooling_rate=0.99):

    # geração do calendário inicial
    calendario_atual = gerar_calendario(times, horarios)
    custo_atual = calcular_custo(calendario_atual, times, distancias)
    melhor_calendario = calendario_atual
    melhor_custo = custo_atual

    for _ in range(max_iter):
        # trocar jogos aleatórios
        calendario_vizinho = gerar_calendario(times, horarios)
        custo_vizinho = calcular_custo(calendario_vizinho, times, distancias)

        # verificar se a solução vizinha é mais otimizada
        delta_custo = custo_vizinho - custo_atual
        if delta_custo < 0 or random.random() < math.exp(-delta_custo / temp):
            calendario_atual = calendario_vizinho
            custo_atual = custo_vizinho

        # definir a melhor otimização
        if custo_vizinho < melhor_custo:
            melhor_calendario = calendario_vizinho
            melhor_custo = custo_vizinho

        # cooling (resfriamento)
        temp *= cooling_rate

    return melhor_calendario

calendario_otimizado = simulated_annealing(times, horarios, distancias)

# print do calendário
for turno, rodadas in calendario_otimizado.items():
    print(f"\n{turno.capitalize()}:")
    for i, rodada in enumerate(rodadas, 1):
        print(f"Rodada {i}:")
        for jogo in rodada:
            print(f"{jogo['mandante']} x {jogo['visitante']} - Estádio: {jogo['estadio']} - Horário: {jogo['horario']}")
```
